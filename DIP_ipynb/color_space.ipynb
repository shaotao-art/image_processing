## 要点
1. 不同色彩空间的取值范围不同
2. 不同色彩空间表达色彩的能力与范围不同
3. 不同色彩空间的取值取决于三维色彩空间的几何结构（如hsv为一个圆柱体）
4. 不同色彩空间的意义是以什么为标准（RGB使得电脑容易处理，hsv则贴近人的感官）

## lab的数值范围有正有负

```python
"""
lab
    L : lightness 0 - 100
    A : green - red -128 - 128
    B : blue - yellow -128 - 128
"""

img_lab=color.rgb2lab(img_f)
## lab   无法正常显示
plt.subplot(2,2,1)
io.imshow(img_lab)

## l
plt.subplot(2,2,2)
io.imshow(img_lab[:,:,0])
## a
plt.subplot(2,2,3)
io.imshow(img_lab[:,:,1])
## b
plt.subplot(2,2,4)
io.imshow(img_lab[:,:,2])
```

```python
"""
hsv/hsb
    H : hue 0 - 360
    S : saturation -128 - 128
    V : value/brightness -128 - 128
"""
img_hsv=color.rgb2hsv(img_f)
## hsv   无法正常显示
plt.subplot(2,2,1)
io.imshow(img_hsv)

## h
plt.subplot(2,2,2)
io.imshow(img_hsv[:,:,0])
## s
plt.subplot(2,2,3)
io.imshow(img_hsv[:,:,1])
## v
plt.subplot(2,2,4)
io.imshow(img_hsv[:,:,2])
```
