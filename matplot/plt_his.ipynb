```python
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
data=np.random.randint(0,100,(1000))
```

```python
x_lst=[0,10,30,40,70]   #x_lst 为 x 轴的起始点和终点
plt.hist(data,bins=x_lst,edgecolor='black')  #将数据分为 5 份
plt.axvline(30,color="red",label="median")
plt.legend()
```

```python
plt.hist(data,bins=10,edgecolor='black',log="True")  #将数据分为 5 份
```

```python
plt.hist(data,bins=10,edgecolor='black')  #将数据分为 5 份
```
