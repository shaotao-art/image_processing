```python
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
x=np.arange(0,3,0.1)
y1=np.exp(x)
y2=np.ones_like(y1)
y3=x
y4=x**2/2
plt.plot(x,y1,label="y=exp(x)")
plt.plot(x,y2+y3+y4,label="y=1+x+x^2/2")
plt.title("taylor")
plt.legend()
```

```python
fig,axs=plt.subplots(1,2)
axs[0].plot(x,y1,label="y=exp(x)")
axs[1].plot(x,y2+y3+y4,label="y=1+x+x^2/2")
axs[0].legend()
axs[1].legend()
```
